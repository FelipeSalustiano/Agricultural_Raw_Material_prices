```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('agricultural_raw_material.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace(' ', np.nan)
df = df.replace('MAY90', np.nan)
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
lst = [
    "Coarse wool Price",
    "Coarse wool price % Change",
    "Copra Price",
    "Copra price % Change",
    "Cotton Price",
    "Cotton price % Change",
    "Fine wool Price",
    "Fine wool price % Change",
    "Hard log Price",
    "Hard log price % Change",
    "Hard sawnwood Price",
    "Hard sawnwood price % Change",
    "Hide Price",
    "Hide price % change",
    "Plywood Price",
    "Plywood price % Change",
    "Rubber Price",
    "Rubber price % Change",
    "Softlog Price",
    "Softlog price % Change",
    "Soft sawnwood Price",
    "Soft sawnwood price % Change",
    "Wood pulp Price",
    "Wood pulp price % Change"
]

df[lst] = df[lst].replace('', np.nan)
df[lst] = df[lst].astype("float")
df.dtypes
```

```python
df.head()
```

```python
df.Month = pd.to_datetime(df.Month.str.upper(), format='%b%y', yearfirst=False)
df = df.set_index('Month')
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
raw_data = ['Coarse wool Price',
            'Copra Price',
            'Cotton Price',
            'Fine wool Price',
            'Hard log Price',
            'Hard log Price',
            'Hide Price',
            'Plywood Price',
            'Rubber Price',
            'Softlog Price',
            'Soft sawnwood Price',
            'Wood pulp Price'
            ]

corrmat = df[raw_data].corr()
```

```python
fig = plt.figure(figsize=(12, 9))
mask = np.triu(np.ones_like(corrmat, dtype=bool))
sns.heatmap(corrmat, vmax= .8, mask=mask, square=True, annot=True)
plt.show()
```

O mapa de calor mostra a correlaÃ§Ã£o entre as matÃ©rias-primas maior o valor correlacionado maior chance de serem duas matÃ©rias-primas relacionadas, mas nÃ£o necessarioamente

```python
plt.figure(figsize=(30, 15))
changelist=['Copra price % Change',
            'Softlog price % Change',
            'Rubber price % Change',
            'Cotton price % Change',
            'Coarse wool price % Change',
            'Hard log price % Change',
            'Fine wool price % Change',
            'Hard sawnwood price % Change',
            'Soft sawnwood price % Change',
            'Wood pulp price % Change',
            'Plywood price % Change',
            'Hide price % change'
            ]

corrMatrix = df[changelist].corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

A partir deste grÃ¡fico podemos dizer que quase nÃ£o hÃ¡ relaÃ§~Ã§ao entre a variaÃ§Ã£o percentual dos preÃ§os das matÃ©rias-primas
- O valor negativo implica que duas varÃ¡veis estÃ£o negativamnte correlacioanadas (um aumento, outro decrÃ©scimo)
- Zero implica nenhuma correlaÃ§Ã£o
- Caso contrÃ¡rio, maior o valor maior a chance de relaÃ§Ã£o, preÃ§os e seus grÃ¡ficos de porcentagem de mudanÃ§a

**Q1: Descubra a variaÃ§Ã£o normal do preÃ§o de cada matÃ©ria-prima**

```python
for i in range(len(changelist)):
  plt.figure(figsize=(12,12))
  df[changelist[i]].hist(figsize=(11, 9), linewidth=1)
  plt.xlabel('%Change')
  plt.ylabel('count')
  plt.legend(changelist[i: ], loc='upper center', bbox_to_anchor=(1.2,1))
```

Podemos observar que a maioria das matÃ©rias-primas tem porcentagem de mudanÃ§a frequente ideal inferior a 5%

**Q2: Encontre a matÃ©ria-prima que tem o menor preÃ§o ao longo dos anos**  

```python
materialslist = ['Coarse wool Price',
            'Copra Price',
            'Cotton Price',
            'Fine wool Price',
            'Hard log Price',
            'Hard log Price',
            'Hide Price',
            'Plywood Price',
            'Rubber Price',
            'Softlog Price',
            'Soft sawnwood Price',
            'Wood pulp Price'
            ]

plt.figure(figsize=(10, 10))
for i in range(len(materialslist)):
  plt.subplot(4, 3, i+1)
  plt.subplots_adjust(hspace=1, wspace=0.5)
  plt.title(materialslist[i])
  plt.plot(df[materialslist[i]])
  plt.xticks(rotation=90)
plt.suptitle('Raw-Materials price comparision')
```

```python
plt.figure(figsize=(10, 10))
plt.plot(df[['Cotton Price', 'Rubber Price']])
plt.title('Raw-Materials price comparison')
plt.xlabel('years')
plt.ylabel('price')
plt.legend(['Cotton Price', 'Rubber Price'], loc='upper center', bbox_to_anchor=(1.2,1))
```

O **algodÃ£o** Ã© a matÃ©ria-prima de menor preÃ§o nos Ãºltimos anos. Tanto o algodÃ£o como a borracha sÃ£o materiais de baixo preÃ§o

**Q3: Qual a matÃ©ria-prima tem a maior e menor variaÃ§Ã£o de porcentagem de preÃ§o?**


```python
import random as random
```

```python
plt.figure(figsize=(12, 12))
for i in range(len(changelist)):

  r = random.random()
  g = random.random()
  b = random.random()
  color = (r, g, b)
  plt.subplot(4, 3, i+1)
  plt.subplots_adjust(hspace=1, wspace=0.5)
  plt.plot(df[changelist[i]], color=color)
  plt.xticks(rotation=90)
  plt.title(changelist[i])
  plt.xlabel('years')
  plt.ylabel('% Change')
```

Podemos ver a maior porcentagem de mudanÃ§a em mais de 60 para madeira serrada macia e a menor porcentagem de mudanÃ§a Ã© para madeira compensada em menos de 20

**Q4:Encontre as matÃ©rias-primas com mudanÃ§a drÃ¡stica de preÃ§o**

```python
lowlist = ['Cotton Price',
           'Hide Price',
           'Softlog Price',
           'Soft sawnwood Price',
           'Hard log Price',
           'Rubber Price'
           ]

plt.figure(figsize=(12,12))
plt.xlabel('prices')
plt.ylabel('years')
for i in range(len(lowlist)):
  sns.scatterplot(y=df[lowlist[i]], x=df.index);
  plt.legend(lowlist, loc='upper center', bbox_to_anchor=(1.2,1))
```

A MudanÃ§a de preÃ§o Ã© drÃ¡stica para o preÃ§o de tora dura entre materiais de baixa faixa de preÃ§o e entre materiais de alto preÃ§o, sÃ£o os preÃ§os de lÃ£ fina

**Q5: Descobrir a faixa de preÃ§o de matÃ©rias-primas de baixo preÃ§o**

```python
plt.figure(figsize=(12,12))
plt.xlabel('prices')
plt.ylabel('years')
for i in range(len(lowlist)):
  plt.title(lowlist[i])
  plt.subplot(2, 3, i+1)
  plt.subplots_adjust(hspace=0.5)
  sns.boxplot(x=df[lowlist[i]])
```

Inclui: O intervalo interquartil estÃ¡ entre Q3 e Q1 mÃ­nimo, primeiro quartil (Q1), mediana, terceiro quartil (Q3) e mÃ¡ximo e outliers
